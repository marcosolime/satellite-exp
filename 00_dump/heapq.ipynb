Heapq *(module)*
- a heap is a **binary tree** stored inside a list
- it satisfies the heap property
- *the smallest element is always at index 0*

`heapq.heappush(heap, x)`
- it appends `x` at the end of the list
- "bubbles it up" until the heap property is restored
- time complexity: $O(\log n)$

`heapq.heappop(heap)`
- removes and returns the smallest element (element 0)
- moves the last element to the root
- "bubbles it down" to restore the heap order

```python
import heapq

h = []

for x in [7, 3, 10, 5]:
    heapq.heappush(h, x)
    print("After pushing", x, "heap =", h)
```

```python
while h:
    print("Pop =", heapq.heappop(h), "heap becomes", h)
```
