DefaultDict

```python
from collections import defaultdict

def_dic = defaultdict(int)
def_dic[3] += 1 # I can do this

def_dic
```

- when we access a missing key (e.g., `d[x]` where `x` wasn't stored before)
- the dictionary automatically creates the entry
- with a default value equal to `int()`, which is **0**

```python
def_dic[3] += 1           # int key
def_dic['cat'] += 1       # string key
def_dic[(1,2)] += 1       # tuple key
def_dic[3.14] += 1        # float key

def_dic[666] = 'evil'       # we can store non ints

def_dic
```

Dict

```python
dic = dict()
dic[3] += 1
```

With regular dictionaries, if a key is not present, it is not automatically initialized

```python
dic[3] = 0
dic
```

But we can add new keys very easily.
